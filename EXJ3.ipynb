```python

def saisir():
  with open('concours.txt', 'a') as f:
    NCIN = input("Entrez le CIN : ")
    NOM = input("Entrez le nom : ")
    PRENOM = input("Entrez le prénom : ")
    AGE = input("Entrez l age : ")
    DECISION =""
    while (DECISION.upper() != "ADMIS" and DECISION.upper() != "REFUSÉ"):
      DECISION= input("Entrez la décision admis ou refusé : ")

    f.write(f"{NCIN};{NOM};{PRENOM};{AGE};{DECISION}\n")

def admis():
  with open('concours.txt', 'r') as f:
    lines = f.readlines()
    with open('admis.txt', 'a') as f_admis:
      for l in lines:
       champs = l.strip().split(';')
       if len(champs) == 5 and champs[4].lower() == 'admis':
        NCIN, NOM, PRENOM, AGE, DECISION = champs
        f_admis.write(f"{NCIN};{NOM};{PRENOM};{AGE};{DECISION}\n")

def attente():
    with open('admis.txt', 'r') as f_admis:
        lines = f_admis.readlines()
        with open('attente.txt', 'a') as f_attente:
          for l in lines:
            champs = l.strip().split(';')
            NCIN, NOM, PRENOM, AGE, DECISION = champs
            if int(AGE) > 30:
              f_attente.write(f"{NCIN};{NOM};{PRENOM}\n")




def statistiques(dec):
  total_candidats = 0
  candidats_dec = 0
  with open('concours.txt', 'r') as f:
    for l in f:
      NCIN, NOM, PRENOM, AGE, DECISION = l.strip().split(';')
      total_candidats += 1
      if DECISION.upper() == dec.upper():
        candidats_dec += 1
  pourcentage_dec = (candidats_dec / total_candidats) * 100
  return pourcentage_dec

def supprimer():
    with open('admis.txt', 'r') as f_admis:
        lines = f_admis.readlines()
    with open('admis.txt', 'w') as f_admis:
        for line in lines:
            champs = line.strip().split(';')
            if len(champs) == 5 and int(champs[3]) <= 30:
                f_admis.write(line)


#saisir()
#admis()
#attente()
supprimer()
DEC=""
while (DEC.upper() != "ADMIS" and DEC.upper() != "REFUSÉ"):
  DEC= input("Entrez la décision admis ou refusé : ")
statistiques(DEC)
```
