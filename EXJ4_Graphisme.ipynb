```python
#formulaire
def actionEvent () :
    lbl.configure(text = "Vous avez etes = "+ champNom . get ()+ champPrenom . get ()+ champAge . get () )

def deleteText():
      champNom.delete(0,END)
      champPrenom.delete(0,END)
      champAge.delete(0,END)
      lbl.configure(text = " ")


from tkinter import *
f = Tk()
f .geometry( '400x200' )


#Création des widgets
Nom = Label( f , text = 'Saisissez votre nom : ' )
champNom = Entry ( f )
Prenom = Label ( f , text= 'Saisissez votre prénom : ' )
champPrenom = Entry ( f )
Age = Label ( f , text= 'Saisissez votre age : ' )
champAge = Entry ( f )


champNom . bind ( "<Return>" , actionEvent )
champPrenom . bind ( "<Return>" , actionEvent )
Age. bind ( "<Return>" , actionEvent )
lbl = Label ( f , text = " . . . . . ")

#Application de la méthode grid () aux widget
Nom. grid (row=0, column=0)
champNom. grid (row=0, column=1)
Prenom. grid (row=1, column=0)
champPrenom. grid (row=1, column=1)
Age. grid (row=2, column=0)
champAge. grid (row=2, column=1)
lbl .grid (row=4, column=0)
b = Button(f , text = "Submit" , command = actionEvent )
b1 =  Button(f , text = "Annuler" , command = deleteText )
b2 = Button(f , text = "Quitter" , command = quit )
b.grid (row=5, column=0)
b1.grid (row=5, column=1)
b2. grid (row=5, column=2)

f .mainloop ()
```

```python
#calculatrice
def calculate():
    operation = champOp.get()
    result = eval(operation)
    champOp.delete(0, END)
    champOp.insert(0, str(result))


def deleteText():
      champOp.delete(0,END)


from tkinter import *
f = Tk()
f .geometry( '400x200' )


#Création des widgets
operation = Label( f , text = 'Opération: ' )
champOp= Entry ( f )


#Application de la méthode grid () aux widget
operation. grid (row=0, column=0)
champOp. grid (row=0, column=1)

b = Button(f , text = "Calculer" , command = calculate )
b1 =  Button(f , text = "Annuler" , command = deleteText )
b2 = Button(f , text = "Quitter" , command = quit )
b.grid (row=5, column=0)
b1.grid (row=5, column=1)
b2. grid (row=5, column=2)

f .mainloop ()
```

```python
from tkinter import *
from random import randrange

# --- définition des fonctions gestionnaires d'événements : ---
def drawline():
    "Tracé d'une ligne dans le canevas can1"
    global x1, y1, x2, y2, coul
    can1.create_line(x1,y1,x2,y2,width=2,fill=coul)
    # modification des coordonnées pour la ligne suivante :
    y2, y1 = y2+10, y1-10

def drawcircle():
    "Tracé d'un cercle dans le canevas can1"
    global rayon, coul, cercle_id
    centre_x = 100
    centre_y = 100

    # Supprimer le cercle précédent s'il existe
    if cercle_id:
        can1.delete(cercle_id)

    # Dessiner un nouveau cercle
    cercle_id = can1.create_oval(centre_x - rayon, centre_y - rayon, centre_x + rayon, centre_y + rayon, fill=coul)

def changecolor():
    "Changement aléatoire de la couleur du tracé"
    global coul
    pal=['purple','cyan','maroon','green','red','blue','orange','yellow']
    c = randrange(8)         # => génère un nombre aléatoire de 0 à 7
    coul = pal[c]

def changerayon():
    "Changement aléatoire du rayon du cercle"
    global rayon
    rayon = randrange(20, 90)  # => génère un nombre aléatoire de 20 à 90


#------ Programme principal -------

# les variables suivantes seront utilisées de manière globale :
x1, y1, x2, y2 = 10, 190, 190, 10      # coordonnées de la ligne
coul = 'dark green'                     # couleur de la ligne
rayon = 40                              # rayon du cercle initial
cercle_id = None                        # identifiant du cercle actuel

# Création du widget principal ("maître") :
fen1 = Tk()

# création des widgets "esclaves" :
can1 = Canvas(fen1, bg='dark grey', height=200, width=200)
can1.pack(side=LEFT)

bou1 = Button(fen1, text='Quitter', command=fen1.quit)
bou1.pack(side=BOTTOM)

bou2 = Button(fen1, text='Tracer Cercle', command=drawcircle)
bou2.pack()

bou3 = Button(fen1, text='Changer couleur', command=changecolor)
bou3.pack()

bou4 = Button(fen1, text='Changer rayon', command=changerayon)
bou4.pack()

fen1.mainloop()
```
