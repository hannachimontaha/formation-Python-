```python
#comptes bancaires
def ajouter_compte(comptes, cin, nom, prenom, numero_compte, solde):
    compte = {
        'cin': cin,
        'nom': nom,
        'prenom': prenom,
        'numero_compte': numero_compte,
        'solde': solde
    }
    comptes[numero_compte] = compte
    print("Compte ajouté avec succès.")


def afficher_compte(comptes, numero_compte):
    if numero_compte in comptes:
        compte = comptes[numero_compte]
        print("Détails du compte:")
        print("CIN:", compte['cin'])
        print("Nom:", compte['nom'])
        print("Prénom:", compte['prenom'])
        print("Numéro de compte:", compte['numero_compte'])
        print("Solde:", compte['solde'])
    else:
        print("Le compte avec le numéro", numero_compte, "n'existe pas.")

def debiter_compte(comptes, numero_compte, montant):
  if numero_compte in comptes:
    compte = comptes[numero_compte]
    solde_actuel = compte['solde']
    if solde_actuel >= montant:
      compte['solde'] -= montant
      print("Débit de", montant, "effectué avec succès sur le compte", numero_compte)
      print("Nouveau solde:", compte['solde'])
    else:
      print("Solde insuffisant pour effectuer le débit sur le compte", numero_compte)
  else:
    print("Le compte avec le numéro", numero_compte, "n'exis3te pas.")


def crediter_compte(comptes, numero_compte, montant):
  if numero_compte in comptes:
    compte = comptes[numero_compte]
    compte['solde'] += montant
    print("Crédit de", montant, "effectué avec succès sur le compte", numero_compte)
    print("Nouveau solde:", compte['solde'])
  else:
    print("Le compte avec le numéro", numero_compte, "n'existe pas.")



print("Menu")
print("1. Ajouter un compte")
print("2. Afficher un compte")
print("3. Débiter un compte")
print("4. Créditer un compte")
print("5. Quitter")
comptes_bancaires = {}
while True:
  choix = int(input("Entrez votre choix (1-4) : "))
  if choix == 1:
    cin = input("CIN : ")
    nom = input("nom : ")
    prenom = input("prénom : ")
    numero_compte = int(input("numéro du compte : "))
    solde = int(input("solde : "))
    ajouter_compte(comptes_bancaires, cin, nom, prenom, numero_compte, solde)
    print("COMPTES", comptes_bancaires)
  elif choix == 2:
    nmr = int(input("Saisir le numéro du compte à afficher : "))
    afficher_compte(comptes_bancaires,nmr)
  elif choix == 3:
    nmr = int(input("Saisir le numéro du compte: "))
    mt = int(input("Saisir le montant à débiter: "))
    debiter_compte(comptes_bancaires,nmr, mt)
  elif choix == 4:
    nmr = int(input("Saisir le numéro du compte: "))
    mt = int(input("Saisir le montant à créditer: "))
    crediter_compte(comptes_bancaires,nmr, mt)
  elif choix == 5:
    print("Merci d'avoir utilisé notre programme.")
    break
  else:
    print("Choix invalide. Veuillez entrer un nombre entre 1 et 4.")
```
